## Abschnitt 2: Operatoren

Operatoren sind Symbole oder Zeichen, die verwendet werden, um Operationen auf Variablen und Werten auszuführen. In Python gibt es verschiedene Arten von Operatoren, die auf unterschiedliche Datentypen angewendet werden können. Im Folgenden werden die gängigsten Operatoren für die verschiedenen Datentypen erklärt:

Arithmetische Operatoren:
* Addition (+): Führt eine Addition zweier Werte durch.
* Subtraktion (-): Führt eine Subtraktion zweier Werte durch.
* Multiplikation (*): Führt eine Multiplikation zweier Werte durch.
* Division (/): Führt eine Division zweier Werte durch und gibt einen Fließkommawert zurück.
* Ganzzahlige Division (//): Führt eine Division zweier Werte durch und gibt den ganzzahligen Teil des Ergebnisses zurück.
* Modulo (%): Gibt den Rest einer Division zweier Werte zurück.

Vergleichsoperatoren:
* Gleichheit (==): Überprüft, ob zwei Werte gleich sind.
* Ungleichheit (!=): Überprüft, ob zwei Werte ungleich sind.
* Größer als (>): Überprüft, ob ein Wert größer ist als ein anderer Wert.
* Kleiner als (<): Überprüft, ob ein Wert kleiner ist als ein anderer Wert.
* Größer oder gleich (>=): Überprüft, ob ein Wert größer oder gleich einem anderen Wert ist.
* Kleiner oder gleich (<=): Überprüft, ob ein Wert kleiner oder gleich einem anderen Wert ist.


Logische Operatoren:
* Und (and): Überprüft, ob beide Bedingungen wahr sind.
* Oder (or): Überprüft, ob mindestens eine Bedingung wahr ist.
* Nicht (not): Kehrt den Wahrheitswert um.

Zunächst legen für zwei Werte x und y an.

```python
x = 3
y = 4
```

Wir führen verschiedene Berechnungen mit x und y durch


```python
addition = x + y

print("Addition: x addiert mit y ist:", addition, "\n")

subtraktion = x - y
print("Subtraktion: x subtrahiert mit y ist:", subtraktion, "\n")

multiplikation = x * y
print("Multiplikation: x multipliziert mit y ist:", multiplikation, "\n")

division = x / y
print("Division: x geteilt durch y ist:", division, "\n")

potenzieren = x ** y
print("Potenzieren: x hoch y ist:", potenzieren, "\n")
```

Wir führen jetzt verschiedene Prüfungen durch.

```python
print("x ist kleiner als y:", x < y)
print("x ist größer als y:", x > y)
print("x ist gleich y:", x == y)
print("x ist nicht gleich y:", x != y)
print("x ist kleiner oder gleich y:", x <= y)
print("x ist größer oder gleich y:", x >= y)
```

## Abschnitt 3: Bedingungen und Schleifen

In diesem Abschnitt werden wir uns mit Bedingungen und Schleifen in Python befassen. Bedingungen ermöglichen es uns, Entscheidungen basierend auf dem Wert von Variablen oder Ausdrücken zu treffen, während Schleifen es uns ermöglichen, Codeblöcke wiederholt auszuführen. Wir werden verschiedene Arten von Bedingungen und Schleifen kennenlernen, um flexibel mit unterschiedlichen Situationen umgehen zu können.

Bedingungen:
* if-Anweisung: Die if-Anweisung ermöglicht es uns, Codeblöcke auszuführen, wenn eine Bedingung wahr ist.
* else-Anweisung: Die else-Anweisung wird verwendet, um Code auszuführen, wenn die Bedingung der if-Anweisung nicht erfüllt ist.
* elif-Anweisung: Die elif-Anweisung wird verwendet, um mehrere Bedingungen nacheinander zu überprüfen, wenn die vorherigen Bedingungen nicht erfüllt waren.

Schleifen:
* while-Schleife: Die while-Schleife wird verwendet, um einen Codeblock so lange auszuführen, wie eine Bedingung wahr ist.
* for-Schleife: Die for-Schleife wird verwendet, um über eine Sequenz (z.B. Liste, Tupel, Zeichenkette) zu iterieren und den Codeblock für jedes Element auszuführen.

Wir führen ein einfaches Beispiel basierend auf x und y durch.

```python
print("x = ", x, "\n")
print("y = ", y, "\n")

if x > y:
    print("x ist größer als y", "\n")

elif x < y:
    print("x ist kleiner als y", "\n")

else:
    print("x ist identisch mit y", "\n")

# Schleifen wiederholen Codeblöcke mehrmals
print("Die ersten 5 natürlichen Zahlen sind:")
for i in range(1, 6):
    print(i)
```

## Abschnitt 4: Funktionen

Funktionen sind benannte Code-Blöcke, die eine bestimmte Aufgabe erfüllen.
Sie werden durch den Schlüsselbegriff **def** in Python definiert und können mit einem Namen versehen werden, um sie später wiederzuverwenden.

Wir können Funktionen definieren, indem wir den Namen der Funktion angeben, gefolgt von Klammern und einem Doppelpunkt. Beim Aufruf einer Funktion geben wir den Funktionsnamen gefolgt von Klammern an, optional mit Argumenten, die der Funktion übergeben werden sollen.

Funktionen können Argumente akzeptieren, die es ermöglichen, Daten an die Funktion zu übergeben, die sie dann verarbeiten kann. Rückgabewerte ermöglichen es einer Funktion, ein Ergebnis oder eine Verarbeitung zurückzugeben, die dann von der aufrufenden Stelle verwendet werden kann.

Gängige Funktionstypen in Python sind:
* Eingebaute Funktionen: Diese sind in Python vordefiniert und können direkt verwendet werden, z.B. print(), len(), range(), usw.
* Benutzerdefinierte Funktionen: Diese werden vom Benutzer erstellt, um spezifische Aufgaben zu erfüllen und können nach Bedarf wiederverwendet werden.
* Lambda-Funktionen: Auch als anonyme Funktionen bekannt, sind sie kurze, einzeilige Funktionen, die ohne Namen definiert werden können und oft in Verbindung mit Funktionen wie map(), filter(), reduce() verwendet werden.
* Rekursive Funktionen: Dies sind Funktionen, die sich selbst aufrufen, um eine Aufgabe zu lösen, indem sie sich wiederholt aufrufen, bis eine Basisbedingung erfüllt ist.

Wir bauen jetzt eine eigene Funktion für einen Blitzer.

```python
#Diese Funktion überprüft, ob die Geschwindigkeit eines Fahrzeugs die zulässige Höchstgeschwindigkeit in einer Tempozone überschreitet.
def Blitzer(Fahrzeug, Geschwindigkeit, Tempozone):
    """
    Parameter:
    - Fahrzeug: Gibt die Bezeichnung des Fahrzeugs an.
    - Geschwindigkeit: Die Geschwindigkeit des Fahrzeugs in km/h.
    - Tempozone: Die zulässige Höchstgeschwindigkeit in der Tempozone in km/h.

    Rückgabewert:
    - True, wenn die Geschwindigkeit die zulässige Höchstgeschwindigkeit überschreitet.
    - False, wenn die Geschwindigkeit innerhalb der zulässigen Höchstgeschwindigkeit liegt.
    """

    if Geschwindigkeit > Tempozone:
        return "Achtung! Das Fahrzeug {} hat die zulässige Höchstgeschwindigkeit in dieser Tempozone überschritten.".format(Fahrzeug)
    else:
        return "Das Fahrzeug {} fährt innerhalb der zulässigen Höchstgeschwindigkeit in dieser Tempozone.".format(Fahrzeug)
```

Wir testen die Funktion "Blitzer" anhand eines fiktiven Beispiels, einem schnellen Audi RS7 und einem langsamen VW Golf 5. Wir geben die Geschwindigkeiten der beiden Fahrzeuge sowie die zulässige Höchstgeschwindigkeit in der Tempozone an und überprüfen, ob die Funktion korrekt erkennt, ob die Geschwindigkeit jedes Fahrzeugs die zulässige Höchstgeschwindigkeit überschreitet oder nicht.

```python
AlterPostweg = 50  # Zulässige Höchstgeschwindigkeit in der Tempozone in km/h

v_AudiRS7 = 80  # Geschwindigkeit des Fahrzeugs in km/h
v_Golf5 = 45  # Geschwindigkeit des Fahrzeugs in km/h


ergebnis_audi = Blitzer("Audi RS7", v_AudiRS7, AlterPostweg)
ergebnis_golf = Blitzer("VW Golf 5", v_Golf5, AlterPostweg)

print(ergebnis_audi)
print( "\n")
print(ergebnis_golf)
```

Beispiel: Lernkurve von Studierenden im Python-Crash-Kurs

Um die Leistungssteigerung über die Zeit zu visualisieren, können wir die Lernkurve mit Matplotlib darstellen. Angenommen, wir haben eine Gruppe von Studierenden, die Python lernen, und wir möchten verfolgen, wie sich ihre Fähigkeiten verbessern, je mehr Zeit sie mit dem Lernen verbringen. Da wir einen nicht-linearen Zusammenhang unterstellen, der mit über Numpy importiert werden kann. Wir erstellen ein Linien-Diagramm, das die Beziehung zwischen den gelernten Stunden und den Python-Fähigkeiten zeigt. Mit Matplotlib können wir das Diagramm beschriften, anpassen und schließlich anzeigen, um die Lernkurve der Studierenden zu visualisieren.

```python
import numpy as np  # Importiere die NumPy-Bibliothek und verkürze sie mit dem Alias "np"

import matplotlib.pyplot as plt  # Importiere die Matplotlib-Bibliothek und verkürze sie mit dem Alias "plt"

# Erzeuge eine Liste von 100 gleichmäßig verteilten Werten zwischen 0 und 5, um die Anzahl der gelernten Stunden darzustellen
stunden_gelernt = np.linspace(0, 5, 100)

# Definiere die logistische Funktion, um den nichtlinearen Zusammenhang zwischen gelernten Stunden und Python-Fähigkeiten darzustellen
# Die logistische Funktion wird als 100 / (1 + e^(-3*(x-2))) definiert, wobei x die Anzahl der gelernten Stunden ist
# e ist die Eulersche Zahl und wird mit np.exp() berechnet
# Der Parameter 3 steuert die Steilheit der Kurve und der Parameter 2 verschiebt die Kurve entlang der x-Achse
python_fähigkeiten = 100 / (1 + np.exp(-3*(stunden_gelernt-2)))

plt.plot(stunden_gelernt, python_fähigkeiten, linestyle='-') # Erstelle ein Linien-Diagramm, um die Lernkurve der Studierenden darzustellen

plt.title('Lernkurve von Studierenden in Python') # Beschrifte das Diagramm mit einem Titel
plt.xlabel('Stunden gelernt') # Beschrifte die x-Achse mit "Stunden gelernt"
plt.ylabel('Python-Fähigkeiten') # Beschrifte die y-Achse mit "Python-Fähigkeiten"
plt.grid(True) # Aktiviere das Gitter im Diagramm, um die Orientierung zu erleichtern
plt.show() # Zeige das Diagramm an
```

### Seaborn: Eine Bibliothek für die (erweiterte) statistische Datenvisualisierung

Seaborn ist eine Python-Bibliothek für die statistische Datenvisualisierung, die auf Matplotlib aufbaut und dessen Funktionalität erweitert. Sie bietet einfache Schnittstellen für komplexe Visualisierungen und ermöglicht es, ästhetisch ansprechende Diagramme mit wenigen Codezeilen zu erstellen. Seaborn ist besonders nützlich für die Erstellung von Heatmaps, Verteilungsplots, Boxplots und anderen fortgeschrittenen Visualisierungen.

Hier ist ein Beispiel, wie man mit Seaborn ein Jointplot-Diagramm erstellt, das die Beziehung zwischen Daten darstellt. Hierfür verwenden wir den integrierten Datensatz "penguins", der die Länge der Flossen in Verhältnis zur Schnabellänge für verschiedene Pinguinarten setzt und visualisieren diesen Zusammenhang:

```python
import seaborn as sns            # Importieren der Seaborn-Bibliothek für die Datenvisualisierung
import matplotlib.pyplot as plt  # Importieren der Matplotlib-Bibliothek für die Erstellung von Diagrammen

penguins = sns.load_dataset("penguins") # Laden des integrierten Datensatzes "penguins" von Seaborn


# Erstellen eines Jointplot-Diagramms
jointplot = sns.jointplot(              # Ein Jointplot zeigt die Verteilung zweier Variablen sowie deren Beziehung zueinander an.
    data=penguins,
    x="flipper_length_mm",  # x und y geben die Variablen an, die im Diagramm dargestellt werden sollen.
    y="bill_length_mm",     # x und y geben die Variablen an, die im Diagramm dargestellt werden sollen.
    hue="species",          # hue definiert eine optionale Variable zur Unterscheidung der Kategorien.
                            )

# Titel für das Diagramm hinzufügen
plt.subplots_adjust(top=0.9)   # Einstellen des Abstands für den Titel, um Platz zu schaffen
jointplot.fig.suptitle("Beziehung zwischen Flossen- und Schnabellänge bei Pinguinen") # Hinzufügen des Titels zum Diagramm

# Achsenbeschriftungen hinzufügen
jointplot.set_axis_labels("Flossenlänge (mm)", "Schnabellänge (mm)")

# Legende in Deutsch hinzufügen
jointplot.ax_joint.legend(title="Pinguinart")

#Optionaler Code Teil um Aus Colab Diagramme zu exportieren:
from google.colab import files      # Importieren der Funktion "files" aus dem Modul "google.colab"
plt.savefig("penguin_plot.png")     # Speichern des Diagramms unter dem Dateinamen "penguin_plot.png" im Google-Drive-Verzeichnis
files.download("penguin_plot.png")  # Herunterladen der PNG-Datei aus Colab

# Diagramm anzeigen
plt.show()
```

### Exkurs Day-Ahead-Strommarkt:
Unter dem Begriff Day-Ahead-Stromhandel versteht man den Handel von Strom für den folgenden Tag, der an der EPEX Spot in Paris stattfindet (Spotmarkt der European Power Exchange). Der Day-Ahead Strommarkt in der Europäischen Union (EU) ist ein wichtiger Bestandteil des europäischen Energiesystems. Auf diesem Markt werden Strommengen für den nächsten Tag gehandelt, wobei Angebot und Nachfrage für jede Stunde am Tag die Preise bestimmen.
* [Weiterführende Informationen zum Day-Ahead-Handel](https://www.next-kraftwerke.de/wissen/day-ahead-handel#entwicklung-des-day-ahead-brsenmarktes
)

Ein zentrales Konzept im Day-Ahead Markt ist die Merit-Order, insbesondere in Deutschland. Die Merit-Order ist eine Rangfolge der Stromerzeugungseinheiten nach ihren variablen Kosten bzw. Grenzkosten der Bereitstellung weiterer Stromvolumina. Die Einheiten mit den niedrigsten variablen Kosten werden zuerst eingesetzt, um die Nachfrage zu decken. Dies bedeutet, dass erneuerbare Energien und entsprechende Kraftwerke mit niedrigen variablen Kosten bevorzugt werden und teurere, z.B. mit Kohle und Erdgasbetriebene Kraftwerke erst bei höherer Stromnachfrage eingesetzt werden, um die verbleibende Nachfrage zu decken.

Die Merit-Order in Deutschland hat eine besondere Bedeutung insbesondere bei einem hohen Anteil erneuerbarer Energien im deutschen Energiemix. Durch die bevorzugte Einspeisung von erneuerbaren Energien werden die variablen Kosten gesenkt und der Einsatz konventioneller Kraftwerke reduziert, was sich wiederum auf den Strompreis auswirkt und sogar zu negativen Preisen führen kann.

### Berechnung der CO2-Intentsität im Strommix für das Jahr 2023
Als nächtes möchten wir die CO2-Intentsität im Strommix für das Jahr 2023 bestimmen. Hierfür benötigen wir die Dateien "Emission_factors.csv" und "Generation_per_production_type_2023.csv" aus den Kursunterlagen. Wir erstellen für beide Datenquellen jeweils ein Dataframe.  

```python
# Wir importieren die benötigten Python-Bibliotheken
from google.colab import drive  # Importiere die drive-Funktion von google.colab, um auf Google Drive zuzugreifen
import pandas as pd  # Importiere die pandas-Bibliothek und verkürze sie als "pd", um Daten zu laden und zu verarbeiten

# Mounten des Google Drive. Mit dieser Zeile verbinden wir unser Google Drive-Konto mit Google Colab, um auf unsere Dateien zuzugreifen
drive.mount('/content/drive')

# Dateipfad zur CSV-Datei im Google Drive-Ordner. Hier definieren wir den Dateipfad zu unserer CSV-Datei auf Google Drive, damit wir sie später laden können
file_path_Emission_factors = '/content/drive/MyDrive/FIM_Python_Crash_Kurs/Emission_factors.csv'
file_path_Generation_per_production_type_2023 = '/content/drive/MyDrive/FIM_Python_Crash_Kurs/Generation_per_production_type_2023.csv'

# Einlesen der CSV-Datei in einen DataFrame. Hier verwenden wir die read_csv-Funktion von pandas, um die CSV-Datei in einen DataFrame zu laden. Der DataFrame ist eine Datenstruktur, die Tabellen ähnelt und uns ermöglicht, mit den Daten zu arbeiten
df_Emission_factors = pd.read_csv(file_path_Emission_factors)
df_Generation_per_production_type_2023 = pd.read_csv(file_path_Generation_per_production_type_2023)

```

```python
# Anzeigen der ersten Zeilen des DataFrames. Hier verwenden wir die .head()-Funktion, um die ersten Zeilen des DataFrames anzuzeigen. Das hilft uns, einen Überblick über die Daten zu bekommen und sicherzustellen, dass sie richtig eingelesen wurden
df_Emission_factors.head()
```

```python
# Anzeigen der ersten Zeilen des DataFrames. Hier verwenden wir die .head()-Funktion, um die ersten Zeilen des DataFrames anzuzeigen. Das hilft uns, einen Überblick über die Daten zu bekommen und sicherzustellen, dass sie richtig eingelesen wurden
df_Generation_per_production_type_2023.head()
```

```python
# Erstellen eines neuen DataFrame für die durchschnittlichen Emissionsfaktoren
df_Average_Emission_factor = df_Generation_per_production_type_2023[['Period', 'Date', 'Time']].copy()

# Hinzufügen einer leeren Spalte 'Average_Emission_factor'
df_Average_Emission_factor['Average_Emission_factor'] = 0.0

# Anzeigen des neuen DataFrames
print(df_Average_Emission_factor)
```

```python
# Berechnung der durchschnittlichen Emissionsfaktoren für jede Zeile
for index, row in df_Average_Emission_factor.iterrows():
    # Initialisierung der Variablen für die Summen
    sum_power = 0  # Summe der Leistungen für jede viertelstunde
    sum_emission_factor = 0  # Summe der gewichteten Emissionsfaktoren für jede viertelstunde

    # Durchlaufen der Spalten E bis Y und Berechnung der gewichteten Summe
    for column in df_Generation_per_production_type_2023.columns:
        if column not in ['Period', 'Date', 'Time']:  # Die Spalten 'Period', 'Date' und 'Time' überspringen
            power = df_Generation_per_production_type_2023.at[index, column]  # Leistung des aktuellen Kraftwerkstyps für die aktuelle Zeile
            emission_factor = df_Emission_factors.at[0, column]  # Emissionsfaktor des aktuellen Kraftwerkstyps
            sum_power += power  # Summe der Leistungen aktualisieren
            sum_emission_factor += power * emission_factor  # Gewichtete Summe der Emissionsfaktoren aktualisieren

    # Berechnung des durchschnittlichen Emissionsfaktors für die aktuelle Zeile
    if sum_power != 0:  # Prüfen, ob durch Null geteilt wird, um Division durch Null zu vermeiden
        average_emission_factor = sum_emission_factor / sum_power  # Durchschnittlicher Emissionsfaktor berechnen
    else:
        average_emission_factor = 0  # Wenn die Summe der Leistungen Null ist, ist der durchschnittliche Emissionsfaktor ebenfalls Null

    # Aktualisierung des Werts in der Spalte "Average_Emission_factor" im DataFrame df_Average_Emission_factor
    df_Average_Emission_factor.at[index, 'Average_Emission_factor'] = average_emission_factor  # Durchschnittlichen Emissionsfaktor für die aktuelle Zeile aktualisieren

# Anzeigen des aktualisierten DataFrames
print(df_Average_Emission_factor)  # Ausgabe des DataFrames df_Average_Emission_factor, der die berechneten durchschnittlichen Emissionsfaktoren enthält
```

### Korrelationsanalyse der Strompreise und CO2-Intentsität im Strommix für die Jahre 2015 - 2023

Im Folgenden  werden wir die Beziehung zwischen Day-Ahead-Strompreisen und der CO2-Intensität im Strommix untersuchen. Strompreise und CO2-Intensität sind wichtige Indikatoren für die Nachhaltigkeit und Wirtschaftlichkeit der Stromerzeugung. Eine Korrelationsanalyse ermöglicht es uns, festzustellen, ob es eine statistisch signifikante Beziehung zwischen diesen beiden Variablen gibt.Wir werden Daten von 2015 bis 2023 verwenden, um langfristige Trends und Muster zu identifizieren.

Hierfür benötigen wir die Datei "Correlation_Day_Ahead_emission_factor.csv" aus den Kursunterlagen. Wir erstellen für die Datenquelle zunächst ein Dataframe.  


```python
from google.colab import drive
import pandas as pd

# Mounten des Google Drive
drive.mount('/content/drive')

# Dateipfad zur CSV-Datei im Google Drive-Ordner
file_path_Correlation_Day_Ahead_emission_factor = '/content/drive/MyDrive/FIM_Python_Crash_Kurs/Correlation_Day_Ahead_emission_factor.csv'

# Einlesen der CSV-Datei in einen DataFrame
df_Correlation_Day_Ahead_emission_factor = pd.read_csv(file_path_Correlation_Day_Ahead_emission_factor)

# Anzeigen der ersten Zeilen des DataFrames
df_Correlation_Day_Ahead_emission_factor.head()
```

Nun möchten wir die Korrelation zwischen den Zeitreihen für jedes Jahr berechnen. Hierfür legen wir einen neuen Dataframe an.

```python
# Erstellen eines neuen DataFrames für die Korrelationen für jedes Jahr

# Wir definieren eine Liste der Jahre von 2015 bis 2023 als Strings
years = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Definieren der Spaltenbezeichnungen für den DataFrame, hier 'Year' für das Jahr und 'Correlation' für den Korrelationswert
columns = ['Year', 'Correlation']

# Erstellen des neuen DataFrames mit 'Year' und 'Correlation' als Spaltenbezeichnungen.
#Wir verwenden die Methode pd.DataFrame(), um ein neues DataFrame zu erstellen.
#'Correlation' wird als Spaltenname für eine weitere Spalte hinzugefügt, aber es sind noch keine Korrelationswerte vorhanden, daher fügen wir vorerst eine leere Liste hinzu.
#Schließlich erstellen wir das DataFrame mit den Spaltenbezeichnungen 'Year' und 'Correlation'
df_yearly_Correlation_Day_Ahead_emission_factor = pd.DataFrame({'Year': years}, columns=columns)

# df_yearly_Correlation_Day_Ahead_emission_factor.head() zeigt die ersten fünf Zeilen des DataFrames an, um sicherzustellen, dass der DataFrames wie erwartet erstellt wurde.
df_yearly_Correlation_Day_Ahead_emission_factor.head()
```

Jetzt berechnen wir die Korrelationen und übergeben diese dem Dataframe. Um die Korrelationen zu berechnen und sie dem DataFrame zuzuweisen, können wir die Pandas-Methode corr() verwenden.

```python
# Iterieren über die Jahre und Berechnen der Korrelationen für jedes Jahr
for year in years:
    # Auswählen der Spalten für das aktuelle Jahr
    day_ahead_prices = f'Day-ahead_Price_{year}_[EUR/MWh]'  # Erstelle den Spaltennamen für den Day-ahead-Strompreis des aktuellen Jahres. Das 'f' vor dem String ermöglicht die Einbettung von Variablen innerhalb des Strings.
    emissions = f'Emissions_{year}_[g/kWh]'  # Erstelle den Spaltennamen für die Emissionen des aktuellen Jahres. Das 'f' vor dem String ermöglicht die Einbettung von Variablen innerhalb des Strings.
    data_year = df_Correlation_Day_Ahead_emission_factor[[day_ahead_prices, emissions]] # Wähle die beiden relevanten Spalten für das aktuelle Jahr aus

    # Berechnen der Korrelation zwischen den beiden ausgewählten Spalten
    correlation = data_year[day_ahead_prices].corr(data_year[emissions])  # Berechne die Korrelation zwischen den Day-ahead-Strompreisen und den Emissionen mit der Pandas .corr() Methode.

    # Aktualisieren des entsprechenden Werts im DataFrame
    df_yearly_Correlation_Day_Ahead_emission_factor.loc[df_yearly_Correlation_Day_Ahead_emission_factor['Year'] == year, 'Correlation'] = correlation  # Aktualisiere den Wert der Korrelations-Spalte für das aktuelle Jahr im DataFrame

# Anzeigen des aktualisierten DataFrames mit den Korrelationen
print(df_yearly_Correlation_Day_Ahead_emission_factor)  # Gebe den DataFrame mit den berechneten Korrelationen aus
```

Schließlich möchten wir über   Seaborn Jointplots der Korrelationsanalyse für jedes Jahr erstellen, um die Korrelationen zwischen den Day-ahead-Strompreisen und den Emissionen über die Jahre hinweg visuell zu untersuchen. Durch diese Visualisierung können wir Trends und Muster in den Daten erkennen und Rückschlüsse auf die Beziehung zwischen Strompreisen und Emissionen ziehen.

```python
# @title
import pandas as pd              # Importiere die pandas-Bibliothek und verkürze sie als "pd", um Daten zu laden und zu verarbeiten
import seaborn as sns            # Importieren der Seaborn-Bibliothek für die Datenvisualisierung
import matplotlib.pyplot as plt  # Importieren der Matplotlib-Bibliothek für die Erstellung von Diagrammen

# Liste der Jahre, für die Jointplots erstellt werden sollen
years = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Iteration über jedes Jahr in der Liste
for year in years:
    # Wähle die Spalten für Day-ahead-Preise und Emissionen für das aktuelle Jahr aus
    day_ahead_prices = f'Day-ahead_Price_{year}_[EUR/MWh]'  # Das 'f' vor dem String ermöglicht die Einbettung von Variablen innerhalb des Strings.
    emissions = f'Emissions_{year}_[g/kWh]'                 # Das 'f' vor dem String ermöglicht die Einbettung von Variablen innerhalb des Strings.

    # Auswahl der relevanten Daten für das aktuelle Jahr
    data_year = df_Correlation_Day_Ahead_emission_factor[[day_ahead_prices, emissions]]

    # Erstellen eines Jointplots mit seaborn
    plt.figure(figsize=(8, 6))  # Festlegen der Größe des Diagramms (Breite x Höhe)
    plot = sns.jointplot(
        data=data_year,  # Daten für das Diagramm
        x=day_ahead_prices,  # Variable auf der x-Achse
        y=emissions,  # Variable auf der y-Achse
        #ratio=5,  # Seitenverhältnis der Joint-Axes
        s=1.5,  # Größe der Punkte im Scatterplot
        color='#B10034', # Farbe der Punkte, kann z.B. auch  "black" lauten
        alpha=1,  # Transparenz der Punkte, der mögliche Wertebereich für alpha liegt zwischen 0 und 1, wobei ein Wert von 1 bedeutet, dass die Punkte vollständig undurchsichtig sind, und ein Wert von 0 bedeutet, dass die Punkte vollständig transparent sind.
        kind='scatter',  # Art des Diagramms (Scatterplot)
        #marginal_kws=dict(bins=50, fill=True)  # Optionen für die Marginaldiagramme (Histogramme)
    )

    # Einstellen der Begrenzungen der Marginaldiagramme
    plot.ax_marg_x.set_xlim(-200, 900)  # Begrenzung der x-Achse des Marginaldiagramms
    plot.ax_marg_y.set_ylim(0, 700)  # Begrenzung der y-Achse des Marginaldiagramms

    #Optionaler Code Teil um Aus Colab Diagramme zu exportieren:
    from google.colab import files      # Importieren der Funktion "files" aus dem Modul "google.colab"
    plt.savefig(f'Correlation_Day_Ahead_emission_factor_{year}.png')  # Das 'f' vor dem String ermöglicht die Einbettung von Variablen innerhalb des Strings. Speichern des Diagramms im Google-Drive-Verzeichnis
    files.download(f'Correlation_Day_Ahead_emission_factor_{year}.png')  # Herunterladen der PNG-Datei aus Colab

    plt.show()  # Anzeigen des Diagramms
```
